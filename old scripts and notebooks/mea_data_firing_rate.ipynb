```python
rate, time = mea_data.get_rate_total(stride=1)

#rate, time = rate[:200000], time[:200000]

sigma = np.std(rate)
mean = np.mean(rate)
```

```python
from scipy.signal import find_peaks

peaks, props = find_peaks(rate, height=3*sigma, distance=5000)

out_of_burst = np.argwhere(rate < .5*sigma)[:,0]

i_lefts = []
i_rights = []

arr = np.array([-5, -3, 0.2, 0.5, 3, 1.5])

burst_durations = []

for peak in peaks:
    left = peak - out_of_burst
    left_indices = np.where(left > 0)[0]
    i_left = left_indices[np.argmin(left[left_indices])]

    right = out_of_burst - peak
    right_indices = np.where(right > 0)[0]
    i_right = right_indices[np.argmin(right[right_indices])]

    i_lefts.append(out_of_burst[i_left])
    i_rights.append(out_of_burst[i_right])
    burst_durations.append((out_of_burst[i_right]-out_of_burst[i_left])/mea_data.sample_rate)
```

```python
plt.hist(burst_durations, bins=20, edgecolor='k')
plt.xlabel('Burst duration [s]')
plt.ylabel('Counts [-]')
plt.show()
```

```python
_, ax1 = plt.subplots(figsize=(20, 8))

plt.title('Neuron Activity over time')

ax1.set_ylabel('Unit number [-]')

for i, row in enumerate(mea_data.trains_binary):
    if i == 0:
        label = 'Firings'
    else:
        label = None
    train = np.argwhere(row)[:,0]
    ax1.scatter(train/mea_data.sample_rate, 
                np.full_like(train, mea_data.unit_numbers[i]), 
                s=.07, color='k', label=label)


ax2 = ax1.twinx() 
for j in range(len(i_lefts)):
    ax2.plot(time[i_lefts[j]:i_rights[j]], rate[i_lefts[j]:i_rights[j]], c='r')
    if j == 0:
        ax2.plot(time[:i_lefts[j]], rate[:i_lefts[j]], c='k')
    elif j == len(i_lefts)-1:
        ax2.plot(time[i_rights[j]:], rate[i_rights[j]:], c='k')
    else:
        ax2.plot(time[i_rights[j-1]:i_lefts[j]], rate[i_rights[j-1]:i_lefts[j]], c='k')

ax2.set_ylabel('Firing rate [Hz]', rotation=270)
plt.xlabel('Time [s]')
plt.xlim(0, 20)
ax2.set_ylim(0, np.max(rate))
ax1.set_ylim(0, mea_data.N_units)
ax2.hlines(3*sigma, 0, 20, linestyle='--', color='r')
ax2.hlines(0.5*sigma, 0, 20, linestyle='--', color='r')
ax2.set_yticks([0, 0.5*sigma, 3*sigma, 8000, 4000], [0, r'$\sigma/2$', r'$3\sigma$', r'$8000$', r'$4000$'], fontsize=14)
plt.show()

```
