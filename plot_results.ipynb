```python
import numpy as np
import pickle
import glob
import matplotlib.pyplot as plt
```

```python
base_path = 'C:\\Users\\bow-lab\\Documents\\Code\\spike_train_analysis\\results_25286'

runs = ['A_1', 'B_1', 'A_2', 'B_2']

files = []

for run in runs:
    files.append(f'{base_path}_{run}.pkl')
```

```python
files = ['C:\\Users\\bow-lab\\Documents\\Code\\spike_train_analysis\\results_25286_A_1.pkl',
          'C:\\Users\\bow-lab\\Documents\\Code\\spike_train_analysis\\results_25286_B_1.pkl',
 'C:\\Users\\bow-lab\\Documents\\Code\\spike_train_analysis\\results_25286_A_2.pkl',
 'C:\\Users\\bow-lab\\Documents\\Code\\spike_train_analysis\\results_25286_B_2.pkl']
```

```python
fig, axs1 = plt.subplots(1, 3, figsize=(18, 5))

fig, axs2 = plt.subplots(1, 3, figsize=(18, 5))

for file, run in zip(files, runs):
    with open(file, 'rb') as file:
        data = pickle.load(file)
        pressures = []
        corr = []
        sc = []
        dt = []
        rates = []
        for d in data:
            print(d)
            pressures.append(d['pressure'])
            corr.append(d['correlation'][0][3])
            sc.append(d['SC_max'])
            dt.append(d['SC_dt_max'])
            rates.append(np.mean(d['rate_total']))
        
        axs1[0].plot(pressures, corr, label=run, marker='s')
        axs1[0].legend()
        axs1[0].set_xlabel('pressure')
        axs1[0].set_ylabel('correlation')

        axs1[1].plot(pressures, sc, label=run, marker='s')
        axs1[1].legend()
        axs1[1].set_xlabel('pressure')
        axs1[1].set_ylabel('spike contrast')
        axs1[1].set_yscale('log')

        #axs1[2].plot(pressures, dt, label=run, marker='s')
        #axs1[2].legend()
        #axs1[2].set_xlabel('pressure')
        #axs1[2].set_ylabel('dt_max spike contrast [s]')

        axs1[2].plot(pressures, rates, label=run, marker='s')
        axs1[2].legend()
        axs1[2].set_xlabel('pressure')
        axs1[2].set_ylabel('average firing rate')

        axs2[0].plot(pressures, corr/corr[0], label=run, marker='s')
        axs2[0].legend()
        axs2[0].set_xlabel('pressure')
        axs2[0].set_ylabel('norm correlation')

        axs2[1].plot(pressures, sc/sc[0], label=run, marker='s')
        axs2[1].legend()
        axs2[1].set_xlabel('pressure')
        axs2[1].set_ylabel('norm spike contrast')
        axs2[1].set_yscale('log')

        #axs2[2].plot(pressures, dt/dt[0], label=run, marker='s')
        #axs2[2].legend()
        #axs2[2].set_xlabel('pressure')
        #axs2[2].set_ylabel('dt_max spike contrast [s]')

        axs2[2].plot(pressures, rates/rates[0], label=run, marker='s')
        axs2[2].legend()
        axs2[2].set_xlabel('pressure')
        axs2[2].set_ylabel('average firing rate')


```
